## **Definición**

La herencia es un proceso mediante el cual se puede crear una clase
hija heredada de una clase padre, compartiendo sus métodos y
atributos.

La clase hija puede sobrescribir los métodos o atributos heredados,
incluso definir unos nuevos.

## **Jerarquía de clases**

Clase 1 (super clase o clase padre) --> Clase 2 (subclase y super clase) --> Clase 3 y Clase 4 (subclases hijas)

## **Importancia de la herencia**

Una de las principales funciones de la herencia es la reutilización de
código al creer objetos similares.

Cualquier clase puede servir como clase padre para construir otras.

# **Polimorfismo**

Este concepto viene de las palabra “poli” muchas y “morfismo”
formas, “muchas formas”.

El polimorfismo en programación se da cuando un objeto cambia de
forma dependiendo del contexto, al hacer esto también cambia su
comportamiento.

El polimorfismo es una característica fundamental en la POO ya que
promueve la flexibilidad, la reutilización de código y la abstracción, lo
que facilita el diseño y la implementación de sistemas complejos.

## **Tipos de Polimorfismo**

- **Polimorfismo de subtipos:** Ocurre cuando una clase hereda de otra y
redefine uno o más métodos de la clase base. Esto permite que los
objetos de la subclase puedan ser tratados como objetos de la clase
base, pero aún así ejecuten el comportamiento específico definido en
la subclase cuando se invocan los métodos.

- **Polimorfismo paramétrico (o genérico):** Se basa en el uso de clases o
métodos genéricos que pueden operar sobre distintos tipos de datos
de manera transparente. Este tipo de polimorfismo permite la
reutilización de código, ya que no es necesario implementar lógica
específica para cada tipo de dato con el que se trabaje.

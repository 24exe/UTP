### Sobrecarga de Operadores

La **sobrecarga de operadores** permite redefinir el comportamiento de los operadores (como `+`, `-`, `*`, etc.) para que funcionen con objetos de una clase personalizada. En Python, esto se logra mediante **métodos especiales** o **"métodos mágicos"**, que tienen nombres específicos como `__add__` para `+`, `__sub__` para `-`, etc.

Por ejemplo, si defines `__add__` en una clase, puedes usar `+` para sumar instancias de esa clase de manera personalizada.

### Sobrecarga en el tipo `str`

Sí, el tipo `str` también usa sobrecarga de operadores en Python. Por ejemplo, la **concatenación** de dos strings con el operador `+` funciona gracias a la sobrecarga del método especial `__add__` en la clase `str`. Además, `str` usa otros métodos especiales como:

- `__mul__`: para repetir la cadena usando el operador `*`.
- `__getitem__`: para acceder a caracteres individuales con `[]`.

Estos métodos permiten que `str` y otros tipos nativos en Python tengan comportamientos específicos con operadores comunes.

````python
def Suma(Op1,Op2):
  return Op1+Op2
def Resta(Op1,Op2):
  return Op1-Op2
def Multi(Op1,Op2):
  return Op1*Op2
def Divi(Op1,Op2):
  return Op1/Op2
````

````python
def Calcular(Oper,Op1,Op2):
  if Oper=="+":
    return Suma(Op1,Op2)
  elif Oper=="-":
    return Resta(Op1,Op2)
  elif Oper=="*":
    return Multi(Op1,Op2)
  elif Oper=="/":
    return Divi(Op1,Op2)
  else:
    return "Error"
````

# **Listas vs Arrays**

## Listas


*   **Tipo de datos:** Las listas son estructuras de datos **heterogéneas**, lo que significa que pueden almacenar elementos de diferentes tipos de datos, como números, cadenas de texto, booleanos e incluso otras listas.
*   **Tamaño dinámico:** Las listas tienen un tamaño **dinámico**, lo que significa que pueden crecer o reducirse según sea necesario. Se pueden agregar o eliminar elementos de una lista sin tener que redimensionarla manualmente.
*   **Operaciones:** Las listas admiten una amplia variedad de operaciones, como agregar elementos (`append`, `insert`), eliminar elementos (`remove`, `pop`), acceder a elementos por índice (`[]`), ordenar elementos (`sort`), etc.
*   **Librería:** Las listas son un tipo de dato **incorporado** en Python, por lo que no es necesario importar ninguna librería para usarlas.

## Arrays

* **Tipo de datos:** Los arrays son estructuras de datos **homogéneas**, lo que significa que todos los elementos de un array deben ser del mismo tipo de datos.

* **Tamaño fijo:** Los arrays tienen un tamaño **fijo**, lo que significa que se debe especificar el tamaño del array al crearlo y no se puede cambiar después.

*  **Operaciones:** Los arrays admiten operaciones similares a las listas, pero generalmente son más eficientes para operaciones numéricas debido a su tipo de datos homogéneo.

* **Librería:** Los arrays se implementan en la librería `array` de Python, por lo que se debe importar esta librería para usarlos.


#**Diccionarios**

*Texto de Gemini*

En Python, un diccionario es una colección de datos desordenada, modificable e indexada. A diferencia de las listas que se indexan por un rango de números, los diccionarios se indexan mediante claves, que pueden ser cualquier tipo de dato inmutable, como cadenas, números o tuplas.

Características principales:

Claves únicas: Cada clave en un diccionario debe ser única. Si intentas agregar una clave que ya existe, se sobrescribirá el valor anterior.
Mutable: Puedes agregar, eliminar y modificar elementos de un diccionario después de su creación.
Desordenado: Los elementos en un diccionario no tienen un orden definido.
Indexado por claves: Para acceder a un valor en un diccionario, se usa su clave correspondiente.
Creación de diccionarios:

Puedes crear un diccionario usando llaves {} y separando las claves y los valores con dos puntos ':'.

Por ejemplo:

```
mi_diccionario = {"nombre": "Juan", "edad": 30, "ciudad": "Madrid"}
```
Acceso a elementos:

Para acceder a un valor en un diccionario, usa la clave entre corchetes []:

```
nombre = mi_diccionario["nombre"]  # nombre será "Juan"
```
Operaciones comunes:

len(diccionario): Devuelve la cantidad de elementos en el diccionario.
diccionario.keys(): Devuelve una lista con todas las claves del diccionario.
diccionario.values(): Devuelve una lista con todos los valores del diccionario.
diccionario.items(): Devuelve una lista de tuplas (clave, valor) para cada elemento del diccionario.

**Ejemplo de uso:**

persona = {"nombre": "Ana", "edad": 25, "profesion": "Ingeniera"}

````python
def validate(values):
  return True if values[0]+values[1]==values[2] else False
````

````python
#esta funcion devuelve un número del largo de la palabra, ej:
#"SEND" se convierte en 9567 (91000 + 5100 + 6*10 + 7).
def get_values(w,v):
  """
  w (una lista de palabras o cadenas de texto)
  v (un diccionario que asigna un valor numérico a cada letra).
  """
  values=[]
  """
  values: Se inicia una lista vacía values donde se almacenarán
  los valores numéricos de cada palabra en w.
  """
  for word in w: #recorre cada palabra en w.
    value=0
    cont=len(word)-1
    """
    -Variable value: Inicializa value en 0 para calcular el valor numérico de
    la palabra actual.
    -Variable cont: Calcula la posición de la primera letra de la palabra
    en base 10. Por ejemplo, si la palabra tiene 4 letras, cont empieza en 3
    para representar la posición de las unidades de mil.
    """
    for l in word: #ciclo interno, recorre cada letra 'l' en la palabra.
      value=value+((10**cont)*v[l])
      cont-=1
    values.append(value)
      #Después de recorrer la palabra completa, el valor numérico value se
      #agrega a la lista values.
  return values
````
